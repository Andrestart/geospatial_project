```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

The coordinates below are the ones from the city center of the cities I got from filtering the database.

```python
coordsf = ['37.778480215129015', '-122.40559118863916']
coordell = ['46.996517823587766', '-120.54550769638885']
coordbro = ['40.692147496548735', '-73.98495174207916']
```

These are the categories for which we have queried:

```python
category = ["vegan_venues", "starbucks", "schools", "train_stations", "basketball_courts" ]
```

# Loading cleaned datasets

```python
dfsanfran = pd.read_csv("../newdata/dfgeosf.csv")
dfell = pd.read_csv('../newdata/dfgeoell.csv')
dfbro = pd.read_csv('../newdata/dfgeobro.csv')
```

# Creating the map for San Francisco

```python
map_sf = folium.Map(location=[coordsf[0],coordsf[1]], zoom_start=13.5)
map_sf;
```

```python
for i,row in dfsanfran.iterrows():
    dic = {"location": [row["lat"], row["lon"]], "tooltip": row["name"]}
    
    if row["category"] == category[0]:
        iconito = Icon(color = "lightgray",
                     prefix="fa",
                     icon="leaf",
                     icon_color="black")
        
    elif row["category"] == category[1]:
        iconito = Icon(color = "green",
                     prefix="fa",
                     icon="coffee",
                     icon_color="black")
        
    elif row["category"] == category[2]:
        iconito = Icon(color = "blue",
                     prefix="fa",
                     icon="graduation-cap",
                     icon_color="black")
        
    elif row["category"] == category[3]:
        iconito = Icon(color = "red",
                     prefix="fa",
                     icon="train",
                     icon_color="black")
        
    elif row["category"] == category[4]:
        iconito = Icon(color = "orange",
                     prefix="fa",
                     icon="globe",
                     icon_color="black")
    mark = Marker(**dic, icon=iconito)
    mark.add_to(map_sf)
map_sf
```
